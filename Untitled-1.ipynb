```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read cscv file
day = pd.read_csv('bike+sharing+dataset/day.csv')
hour = pd.read_csv('bike+sharing+dataset/hour.csv')
```

```python
sns.set_palette("dark")
```

```python
#plot histogram for temp humidity and windspeed side by side using seaborn 
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
#bins = 20, palette = 'deep'
sns.histplot(day['temp'], ax=ax[0], bins=20)
ax[0].set_title('Temperature')
sns.histplot(day['hum'], ax=ax[1], bins=20)
ax[1].set_title('Humidity')
sns.histplot(day['windspeed'], ax=ax[2], bins=20)
ax[2].set_title('Windspeed')
```

```python
#histograms to be plotted in the same figure

#same bins for all histograms

fig, ax = plt.subplots(1, 1, figsize=(11, 5))
 

sns.histplot(day['hum'],  ax=ax, label='Humidity',binwidth=0.021)
sns.histplot(day['windspeed'],  ax=ax, label='Windspeed',binwidth=0.021)
sns.histplot(day['temp'],  ax=ax, label='Temperature',binwidth=0.021)
ax.set_title('Temperature, Humidity and Windspeed')
ax.legend()
plt.show()
```

```python
# plot scatter plot for temp, humidity and windspeed vs cnt using seaborn
fig, ax = plt.subplots(1, 1, figsize=(5, 5))
sns.set_palette('deep')
sns.scatterplot(data=day, x='temp', y='cnt', ax=ax, label='Temperature')
sns.scatterplot(data=day, x='hum', y='cnt', ax=ax, label='Humidity', marker='s')
sns.scatterplot(data=day, x='windspeed', y='cnt', ax=ax, label='Windspeed',marker='^')
ax.set_title('Temperature, Humidity and Windspeed vs Count')
ax.legend()

plt.show()
```

```python
# plot scatter plot for temp, humidity and windspeed vs cnt

fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].scatter(day['temp'], day['cnt'])
ax[0].set_title('Temperature vs cnt')
ax[1].scatter(day['hum'], day['cnt'])
ax[1].set_title('Humidity vs cnt')
ax[2].scatter(day['windspeed'], day['cnt'])
ax[2].set_title('Windspeed vs cnt')
plt.show()
```

```python
#calculate the correlation matrix bw temp, humidity, windspeed and cnt
day[['temp', 'hum', 'windspeed', 'cnt']].corr()
```

```python
#create a bar chart showinmg average cnt for each season
fig, ax = plt.subplots(1, 1, figsize=(5, 5))
sns.color_palette_palette('deep')
sns.barplot(data=day, x='season', y='cnt', ax=ax)
ax.set_title('Average count for each season')
plt.show()
```
